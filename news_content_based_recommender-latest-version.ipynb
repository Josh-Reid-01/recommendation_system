```python
url = "https://rss.app/feeds/qYEIAK4I1h43dPec.xml"

resp=requests.get(url)
soup = BeautifulSoup(resp.content, features="xml")
print(soup.prettify())
```

```python
items = soup.findAll('item')
print(items)
```

```python
news_items = []
for item in items:
    news_item={}
    news_item['title']=item.title.text
    news_item['description']=item.description.text
    news_item['link']=item.link.text
    news_item['pubDate']=item.pubDate.text
    news_items.append(news_item)
news_items[0]
```

```python
data=pd.DataFrame(news_items,columns=['title','description','link','pubDate'])
data
```

```python
feature = data["title"].tolist()
tfidf = text.TfidfVectorizer( stop_words="english")
tfidf_matrix = tfidf.fit_transform(feature)
similarity = cosine_similarity(tfidf_matrix)
```

```python
indices = pd.Series(data.index,data['title']).drop_duplicates()

#result = indices['title'].str.contains(pat = 'Ukraine')
print(type(indices))
#T='Ukraine'
#print (indices[indices.index.str.contains(T)])
    
```

```python
def news_recommendation2(Title, similarity = similarity):
   
    index = indices[indices.index.str.contains(Title)]
    newslink = data[['title', 'link']]
    for j in index:
     similarity_scores = list(enumerate(similarity[j]))
    
     similarity_scores = sorted(similarity_scores,key=lambda x: x[1],reverse=True)
     similarity_scores = similarity_scores[0:20]
    
     newsindices = [i[0] for i in similarity_scores]
     recommend = newslink.iloc[newsindices]
     
    return recommend
     
     
    
    
    
```

```python
entered_title = "Business"
recommendations2 = news_recommendation2(entered_title)
print(recommendations2)
```
